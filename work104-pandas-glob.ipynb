```python
import pandas as pd
import glob
```

```python
glob.glob('./job104/*.txt')
```

```python
data=list()
for filePath in glob.glob('./job104/*.txt'):
    with open(filePath, 'r' , encoding='utf-8') as f:
        tmplist = f.read().split('---split---')
        data.append(tmplist[0:-1])
data
```

```python
columns=['公司名稱','職缺名稱','職缺內容','擅長工具','URL']
columns
```

```python
df = pd.DataFrame(data=data,columns=columns)
df
```

```python
df.to_csv('./job104.csv',index=0,encoding='utf-8-sig')
```
